### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
torch.mm(tensor_A, tensor_B) # torch.mm == torch.matmul
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**. A **transpose** switches the axes or dimensiosn of a given tensor.

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
# The mm op works when tensor_B is transposed
print(f'Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}')
print(f'New shapes: tensor_A = {tensor_A.shape}, tensor_B.T = {tensor_B.T}')
print(f'Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dims must match')
output = torch.mm(tensor_A, tensor_B.T)
print(output)
print(f'\nOutput shape: {output.shape}')
```

## Reshaping, stacking, squeezing, and unsqueezing tensors

* Reshaping - reshape an input tensor to a defined shape
* View - return a view of an input tensor of a certain shape, but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - return a view of the input w/ dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the og tensor)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
x_reshaped
x_squeezed = x_reshaped.squeeze()
```

```python
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
# torch.unsqueeze() - adds single dimension to a target tensor at a specific dim (dimension)
print(f'Previous target: {x_squeezed}')
print(f'Previous shape: {x_squeezed.shape}')

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f'\nNew tensor: {x_unsqueezed}')
print(f'New shape: {x_unsqueezed.shape}')
```

```python
# torch.permute - rearranges the dimensions of target tensor in specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, color_channels]

# Permute og tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f'Previous shape: {x_original.shape}')
print(f'New shape: {x_permuted.shape}')
```

## PyTorch Tensors & NumPy

PyTorch has functionality to interact w/ NumPy.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`


```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: PyTorch reflects np's default datatype of float64
array, tensor
```

```python
# Change the value of array, what will this do to `tensor`?
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what happens to `numpy_tensor`?
tensor = tensor + 1
tensor, numpy_tensor
```

## Reproducability (trying to take random out of random)

In short, how a neural network learns:

`start with random numbers -> tensor ops -> update random numbers to try and make them better representations of the data -> again -> again -> again...`

To reduce the randomness in neural networks and PyTorch, comes the concept of a **random seed**. Essentially what the random seed does is "flavor" the randomness.

```python
import torch

# Create two random tensors
rand_tensor_A = torch.rand(3, 4)
rand_tensor_B = torch.rand(3, 4)

print(rand_tensor_A)
print(rand_tensor_B)
print(rand_tensor_A == rand_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
